```python
#Q.1 Discuss string slicing and provide examples.

#String Slicing:
#--------------

#String slicing means extracting a part of string (i.e. sub-string using 
#indices (positive indexing starts from zero [left to right] whereas negative 
#indexing starts from -1 [right to left])

#example:- suppose we have following string;

string="Bhaskar"

#indexing of above string ==> B  h  a  s  k  a  r
#   positive indexing         0  1  2  3  4  5  6
#   negative indexing        -7 -6 -5 -4 -3 -2 -1  

#syntax for string slicing ==> string_name[start:end:step]
#If I want to access substring 'Bhas' of above string then for this, we 
#can write..

#Using positive index
#--------------------
print(string[0:4])
#the above print statement will print elements of string 'Bhaskar' starting 
#from index zero upto index 3 (i.e. end-1)[end index is excluded]


#Using negative index
#--------------------
print(string[-7:-3])
#this will also fetch me sub-string 'Bhas'


#Case1:
#------
#Only specifying starting index
print(string[3:])

#This will then fetch me entire sub-string starting from index 3 upto last index
#viz. 'skar'

#Case2:
#------
#Only specifying end index

print(string[:-2])
#The output will be 'Bhask' i.e if we do not specify starting index, it fetches
#values from the very first index upto end-1 (-7 to -3).

#Case3:
#------
#Specifying neither starting nor end index

print(string[:])
#The Output will be 'Bhaskar' i.e. in this case complete string from beginning 
#to end will be accessed.

#Case4:
#------
#Slicing with 'step'

print(string[0:6:2])

#The Output will be 'Bak'.
#Explanation:
#------------

# B  h  a  s  k  a  r
# 0  1  2  3  4  5  6
# It will print the 0th index first i.e 'B' 
#After that, we move forward by 2 steps(indices) and we find character 'a'
#after reaching 'a', again we move forward by 2 steps(indices) and this time
#we find character 'k'
#Now since the end index is 6. So, we can only move upto character at index 5.
#i.e. we do not have further indices to move forward by 2 steps.
#Thus, final output is as 'Bak'.

#Case 5:
#-------
#Specifying only step

print(string[::3])
#The output will be 'Bsr'
#i.e. starting from the 0th index till the last index, character at every third
#step has been printed.

#Case 6:
#-------
#Only specifying first index and step
print(string[1::3])
#Output: hk 
#starting from index 1 upto last index, character at every third step has been
#printed

#Case 7:
#-------
#Only specifying end index and step
print(string[:4:2])
#Output: Ba
#starting from 0th index upto index end-1 i.e(4-1 = 3), character at every 2nd 
#step has been printed.

 
```

```python
#Q.2 Explain the key features of list in Python.

#Lists are one of the data structures in Python which are represented using
#square ( '[]' ) brackets.

#The following are important features of lists in Python.

#(1)Ordered collection of elements :
#------------------------------------

# A list is an ordered collection of elements i.e a list retains the order
#of its elements in which they are entered into the list. This property of lists
#allow to access its elements using indexing technique/method.
#Example:
#--------

veggies=['Potato','Brinjal','Pumpkin']

veggies[2] #This will fetch us 'Pumpkin' as it lies at index 2.

#(2) List can store variable data types:
#---------------------------------------

# A list can store elements of different data types like integer,float,string,
#complex numbers etc.

#Example:
random_list=[2020,44.89,'Saksham',2+5j]
random_list[1]

#(3) List can store different Data Structures:
#---------------------------------------------
#We can also store different data structures like sets, tuples and dictionaries
#within a list.

ds_list=[(1,2,3),{1,5,5,2,3},{'name':'Saksham','phone_no':75201,'Age':21}]

ds_list[2]

#Output: {'name': 'Saksham', 'phone_no': 75201, 'Age': 21}

#We can also access specific value from this dictionary as follows:

ds_list[2]['phone_no']

#Output: 75201 which is the value corresponding to key 'phone_no'

#ds_list[1][2]
#this will raise an error beacuse at index 1 there exists a set and elements of
#a set can not be accessed using subscript or index.

ds_list[0][1]
#Output: 2 as at index 0 of the list there exist tuple (1,2,3) and in this 
#tuple element at index 1 is 2.

#(4) Lists are mutable objects:
#-------------------------------
#Lists are mutable in nature i.e once a list is declared its elements
#can be changed later as per the requirement and the changes made are
#reflected in the original list itself.

indian_team=['rohit','virat','hardik','suryakumar']
indian_team[0]='Captain Rohit'#trying to replace element at index 0 viz 'rohit'
#with 'Captain Rohit'
print(indian_team)
#Output:['Captain Rohit', 'virat', 'hardik', 'suryakumar']
#i.e the changes made are visible in the original list indian_team.

#There are various methods(in-built functions) used to manipulate a list.
#Some of them are discussed below:

#(1) append(object):
#--------------------
#append new object at the end of list.takes exactly one argument
indian_team.append('Jadeja')
print(indian_team)
#Output: ['Captain Rohit', 'virat', 'hardik', 'suryakumar', 'Jadeja']

#(2) insert(index,object):
#--------------------------
#inserts a new element in the list before a specifc index.
indian_team.insert(3,'Bumrah')
print(indian_team)
#Output: ['Captain Rohit', 'virat', 'hardik', 'Bumrah', 'suryakumar', 'Jadeja']

#(3) remove(object):
#-------------------
#removes the first occurrence of specified element

list1=[1,2,3,4,2]
list1.remove(2)
print(list1)
#Output: [1, 3, 4, 2]

#(4) extend():
#-------------
# add list to a list
s1=[1,2,3]
s1.extend([4,5,6])
print(s1)
#Output: [1, 2, 3, 4, 5, 6]

#(5) pop():
#----------
#Removes and returns the element at the specified position. If no position 
#is specified, it removes and returns the last element.
my_list = [1, 2, 3]
popped_element = my_list.pop(1)
print(popped_element)  # Output: 2
print(my_list)         # Output: [1, 3]

#(6) clear():
#------------

#Removes all elements from the list.
my_list = [1, 2, 3]
my_list.clear()
print(my_list)  # Output: []

#(7)index():
#-----------

#Returns the index of the first occurrence of a specified value.

my_list = [1, 2, 3]
index_of_2 = my_list.index(2)
print(index_of_2)  # Output: 1

#(8)count():
#-----------

#Returns the number of times a specified value occurs in the list.
my_list = [1, 2, 3, 2, 2]
count_of_2 = my_list.count(2)
print(count_of_2)  # Output: 3

#(9) sort():
#------------
#Sorts the list in ascending order. It can also sort in descending order 
#if specified
my_list = [3, 1, 2]
my_list.sort()
print(my_list)  # Output: [1, 2, 3]

my_list.sort(reverse=True)
print(my_list)  # Output: [3, 2, 1]

#(10)reverse():
#--------------
#Reverses the order of the list.
my_list = [1, 2, 3]
my_list.reverse()
print(my_list)  # Output: [3, 2, 1]

#(5)Lists support concatenation:
#-------------------------------

a1=[1,2,3,4,5,6]
a2=['ram','shyam','mohan','rohan','sohan']
print("list after concatenation:",a1+a2)
#Output: 
#list after concatenation:
#[1,2,3,4,5,6,'ram', 'shyam', 'mohan', 'rohan', 'sohan']

#(6) List Comprehension:
#------------------------
#Lists can be created using list comprehensions, which provide a concise 
#way to generate lists.

squares = [x**2 for x in range(10)]
print(squares)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

#(7) List support Slicing:
#-------------------------
#Lists support slicing, which allows for extracting parts of the list.

my_list = [1, 2, 3, 4, 5]
sub_list = my_list[1:4]
print(sub_list)  # Output: [2, 3, 4]
```

```python
#Q.3 Describe how to access,modify and delete elements in a list with examples.


#Accessing elements in a list:
#==============================

#Since lists are ordered collection of elements, their elements can be accessed
#using indices.

#Example: 
ran_list=[100,273.15,8+2j,'Sarthak',True]

ran_list[0] #100
ran_list[1] #273.15
ran_list[2] #8+2j
ran_list[3] #Sarthak
ran_list[4] #True

#we can also iterate over list of elements using a for loop.

for element in ran_list:
    print(element)
    
#Modifying elements in a list
#============================

#Since lists are mutable objects, their elements can be modify using methods or 
#using indices.

prime_list=[2,3,5,7,11,13,17,19]

prime_list[4]='eleven'# replacing element at index 4 with 'eleven'

print(prime_list)#Output:[2, 3, 5, 7, 'eleven', 13, 17, 19]

prime_list[-1]='nineteen'#replacing the last element of list with 'nineteen'

print(prime_list)#Output:[2, 3, 5, 7, 'eleven', 13, 17, 'nineteen']

#modification using in-built functions
#-------------------------------------
my_list=[1,2,3]

my_list.append(4)
#Adds a new element to the end of the list

print(my_list)#Output :[1,2,3,4]

my_list.insert(0,0)
#inserts new element in a list before specified index
print(my_list)

my_list.extend([5,6,7])
#adds a new iterable list of elements in the list mentioned before dot operator.
print(my_list)#Output:[0, 1, 2, 3, 4, 5, 6, 7]

my_list.reverse()
#reverse the order of elements of a list
print(my_list)#Output:[7, 6, 5, 4, 3, 2, 1, 0]


list_of_words=['zebra','potato','toy','owl','bowl']
list_of_words.sort()
#sort the elements in ascending order.
print(list_of_words)#Output:['bowl', 'owl', 'potato', 'toy', 'zebra']


#Delete elements in a list
#=========================

book_list=['Rich Dad Poor Dad','Atomic Habits','Psychology of Money']

book_list.remove('Rich Dad Poor Dad')
#removes spcified element from the list

print(book_list)#Output:['Atomic Habits', 'Psychology of Money']

book_list.pop()
#deletes the last element from the list by default if no index is specified.
print(book_list)#Output : ['Atomic Habits']

book_list.clear()
#returns an empty list
print(book_list)#Output:[]
```

```python
#Q.4 Compare and contrast tuples and lists with examples.

#Similarities in Tuples and Lists in Python
#+++++++++++++++++++++++++++++++++++++++++++

#(1)Both tuples and list can store elements with different data types.

LIST=[1,1.5,'Arjun',True,2+5j]
print(LIST,type(LIST))
#Output: [1, 1.5, 'Arjun', True, (2+5j)] <class 'list'>

TUPLE=(1,1.5,'Arjun',True,2+5j)
print(TUPLE,type(TUPLE))
#Output: (1, 1.5, 'Arjun', True, (2+5j)) <class 'tuple'>

#(2)Both the tuples and lists can store duplicate values/elements

random_list=[1,2,3,'Saksham','Saksham',1,3,9,9]
print(random_list[3])#Output:Saksham
print(random_list[4])#Output:Saksham
print(random_list)#Output:[1, 2, 3, 'Saksham', 'Saksham', 1, 3, 9, 9]

random_tuple=('Sachin',22,22,55.5,'Raghav')
print(random_tuple[1])#Output: 22
print(random_tuple[2])#Output: 22
print(random_tuple)#Output: ('Sachin', 22, 22, 55.5, 'Raghav')

#(3)Both tuples and lists are ordered collection of elements i.e. both lists
#and tuples retain the order in which the elements are inserted in them.

ek_list=[9,8,7,6,5]
for i in ek_list:
    if(i==5):
        print(i)
        break;
    print(i,end=" --> ")
    
#Output: 9 --> 8 --> 7 --> 6 --> 5

ek_tuple=(-1,-2,-3)
for i in ek_tuple:
    if(i==-3):
        print(i)
        break;
    print(i,end=" --> ")
 #Output : -1 --> -2 --> -3

#(4)Both list and tuples support slicing

random_list=[1,2,3,'Saksham','Saksham',1,3,9,9]
print(random_list[4:8])
#Output:['Saksham', 1, 3, 9]

random_tuple=('Sachin',22,22,55.5,'Raghav')
print(random_tuple[0:3])
#Output:('Sachin', 22, 22)

#Differences between Tuples and Lists in Python:
#+++++++++++++++++++++++++++++++++++++++++++++++

#(1)Tuples are written within small brackets() whereas Lists are written in 
#square bracket[]

a1=[1,2,3]
print(type(a1))
#Output:<class 'list'>
a2=(1,2,3)
print(type(a2))
#Output:<class 'tuple'>

#(2) Lists are mutable whereas Tuples are immutable.

#Mutability of Lists
a1=[1,2,3]
a1[1]=1000
print(a1) # Output: [1,1000,3]

#Immutability of Tuples

a2=(1,2,3)
a2[0]=100#raise error because tuples do not support item assignment.
#print(a2)#Output:(1,2,3)

#tuples are best suited when changes are not supposed to be allowed once values
#are entered.

#(3) Tuples have  fewer methods or in-built functions unlike lists which have
#various methods.

#methods for tuples:
#count(),index()

#methods for Lists:
#pop(),insert(),append(),copy(),index(),clear(),remove(),sort() etc

#(4) Lists because of their dynamic nature consume more memory space whereas 
#tuples take less memory due to their immutability (fixed collection of items)
#Also, tuples have a faster performance as compared to lists due to immutability.
```

```python
#Q.5 Describe the key features of sets and provide examples of their use.

#Sets are one of the data structures used in Python to store values/elements.
#They are represented by curly braces{}

#Key features of sets in Python
#++++++++++++++++++++++++++++++
#(1) Sets are unique collection of elements i.e. 
# a set does not allow duplicate values.if duplicate values are entered
#they are automatically removed when a set is created.

hamara_set={20,30.0,'Twenty',20,30.0,'Twenty'}

print(hamara_set)#Output: {20, 30.0, 'Twenty'}

#(2) Elements in a set can't be accessed or changed using indices.
#sets do not maintain a specific order of elements.
#They are unordered collection collection of elements which do not support 
#indexing.

#print(hamara_set[1])#error: set object is not subscriptable

#hamara_set[2]='apple'#error: set object does not support item assignment.

hamara_set.add('NEW ELEMENT')

print(hamara_set)#Output :  {20, 'NEW ELEMENT', 30.0, 'Twenty'}

hamara_set.add(273.88)

print(hamara_set)#Output : {'NEW ELEMENT', 273.88, 20, 'Twenty', 30.0}

#i.e. no ordering is being followed here.

#(3) Sets are mutable in nature i.e they can be manipulated using
#in-built methods after being created.

#Here are some commonly used methods of sets.

hamara_set.add('in-built func.')
#add new element to a set at a random place.
print(hamara_set)#{'NEW ELEMENT', 'in-built func.', 273.88, 20, 'Twenty', 30.0}

hamara_set.remove('Twenty')
#removes specified element from set.raise error if element not found
print(hamara_set)#{'NEW ELEMENT', 'in-built func.', 273.88, 20, 30.0}

element=hamara_set.pop()
#deletes a random element from the list and return it.
print(element)

#(4) Elements within a set must be of immutable type i.e lists,dictionaries,sets
#can't be placed in a set as it will raise an error but we can store strings,
#and tuples within a set.

#my_set={{'name':'saksham'},2,3}
#print(my_set)# error: unhashable type: 'dict'

#my_set2={1,2,3,[4,5]}
#print(my_set)## error: unhashable type: 'list'


my_set3={1,2,(3,4),'six','seven'}
print(my_set3)#Output: {1, 2, (3, 4), 'seven', 'six'}

#(5) Only sets have methods like union,intersection,difference,
#symmetric difference.The original elements in sets are not modified.

SET_ONE={1,2,3,4}
SET_TWO={3,4,5,6,7}

print(SET_ONE.union(SET_TWO))#{1, 2, 3, 4, 5, 6, 7}
#Combine elements of both sets with common elements included only once.

print(SET_ONE.intersection(SET_TWO))#{3,4}
#returns common elements only from both sets

print(SET_ONE.difference(SET_TWO))#{1,2}
#returned those elements of <<SET_ONE>> which are not in SET_TWO

print(SET_ONE.symmetric_difference(SET_TWO))#{1, 2, 5, 6, 7}
#all the non-common elements from both the sets are returned.


            
```

```python
#Q.7 Describe how to add,modify and delete items in a dictionary with examples.

#Dictionary is a mutable data structure in Python used to store data in the 
#form of key-value pairs.

# Adding elements to a dictionary
#=================================

personal_info={}#intializing an empty dictionary
personal_info["name"]='Saksham Shrivastava'
personal_info['age']=21
personal_info['DOB']='07 November 2003'
personal_info['Gender']='Male'

print(personal_info ,'\n', type(personal_info))

#Output:

#{'name': 'Saksham Shrivastava', 'age': 21, 'DOB': '07 November 2003', 'Gender':
#'Male'} 
#<class 'dict'>

#Elements(key-value pairs) can also be added using update() in-built method
personal_info.update({'city':'Indore','pincode':561022})
print(personal_info)

#{'name': 'Saksham Shrivastava', 'age': 21, 'DOB': '07 November 2003', 
#'Gender': 'Male', 'city': 'Indore', 'pincode': 561022}

#Modify elements in a dictionary
#================================

#It is important to note that keys in dictionaries are not modifiable.
#Only values corresponding to keys can be modified.
#However we can delete a key-value pair as a whole and add new key-value pair.

personal_info['age']=23
personal_info['city']='Bhopal'
print(personal_info)

#Output:
#{'name': 'Saksham Shrivastava', 'age': 23, 'DOB': '07 November 2003', 
#'Gender': 'Male', 'city': 'Bhopal', 'pincode': 561022}

del personal_info['DOB']
#removing key-value pair 'DOB' : '07 November 2003' 
personal_info['Date_of_birth']='04 Oct 2001'
print(personal_info)

#Output:
#{'name': 'Saksham Shrivastava', 'age': 23, 'Gender': 'Male', 'city': 'Bhopal',
#'pincode': 561022, 'Date_of_birth': '04 Oct 2001'}


#Deleting elements from a dictionary:
#====================================
#Elements from a dictionary can be deleted using various methods.

dictionary={1:'saksham',2:'shivam',3:'sarthak'}

#using del statement
#-------------------

del dictionary[2]
print(dictionary)
#Output: {1: 'saksham', 3: 'sarthak'}


#using pop() method
#------------------
#removes key-value pair from dictionary corresponding to key provided in pop()
#and returns value.Raise an error if key not found.
popped_element=dictionary.pop(1)
print(popped_element)
#Output: Saksham
print(dictionary)
#Output:{3: 'sarthak'}


#using clear() method
#---------------------

# Empty dictionary and returns all the key-value pairs of dictionary

bike_models={'Honda':'SP125','Hero':'Xtreme 360 R'}

print(bike_models,type(bike_models))
#Output: {'Honda': 'SP125', 'Hero': 'Xtreme 360 R'} <class 'dict'>
bike_models.clear()
print(bike_models)#Output: {}
```
